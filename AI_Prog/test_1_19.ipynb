```python
def gcd(m, n):
    while m % n != 0:
        m, n = n, m % n
    return n

class Fraction:

    def __init__(self, top, bottom): 
   
        self.num = top
        self.den = bottom
        
    def __str__(self):
        return f"{self.num}/{self.den}"
    
    def __add__(self, other_fraction):
        new_num = self.num*other_fraction.den + self.den*other_fraction.num
        new_den = self.den*other_fraction.den
        common = gcd(new_num, new_den) 
        return Fraction(new_num//common, new_den//common)
    
    def __sub__(self, other_fraction):
        new_num = self.num*other_fraction.den - self.den*other_fraction.num
        new_den = self.den*other_fraction.den
        common = gcd(new_num, new_den) 
        return Fraction(new_num//common, new_den//common)
    
    def __mul__(self, other_fraction):
        new_num = self.num*other_fraction.num
        new_den = self.den*other_fraction.den
        common = gcd(new_num, new_den) 
        return Fraction(new_num//common, new_den//common)
    
    def __truediv__(self, other_fraction):
        new_num = self.num*other_fraction.den
        new_den = self.den*other_fraction.num
        common = gcd(new_num, new_den) 
        return Fraction(new_num//common, new_den//common)
    
    def add(self, other_fraction):
        new_num = self.num*other_fraction.den + self.den*other_fraction.num
        new_den = self.den*other_fraction.den
        common = gcd(new_num, new_den) 
        return Fraction(new_num//common, new_den//common)
    
    def __eq__(self, other_fraction):
        first_num = self.num * other_fraction.den
        second_num = other_fraction.num * self.den
        return first_num == second_num
    
    def __ne__(self, other_fraction):
        first_num = self.num * other_fraction.den
        second_num = other_fraction.num * self.den
        return first_num != second_num

my_fraction = Fraction(3,5)
print(my_fraction, type(my_fraction))
f1 = Fraction(1,4)
f2 = Fraction(1,2)
f3 = f1.add(f2)
f4 = f1 + f2
print(f3)
print(f4)
print(f1 != f2)
f5 = f1 * f2
print(f5)
f6 = f1 - f2
print(f6)
f7 = f1 / f2
print(f7)
```

```python
import numpy as np
#  a y  b
fake_a_b=[3, 76]
# x  y
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# y=ax + b a,b     
def predict(x):
    return fake_a_b[0]*x + fake_a_b[1]

# MSE 
def mse(y, y_hat):
    return ((y - y_hat) ** 2).mean()

# MSE   y     
def mse_val(y, predict_result):
    return mse(np.array(y), np.array(predict_result))

#    
predict_result = []

#  x    predict_result .
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("=%.f, =%.f, =%.f" % (x[i], y[i], predict(x[i])))
    
#  MSE 
print("MSE : " + str(mse_val(predict_result,y)))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# X  Y  .
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

#  .
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.show()

#   x y   .
#(        )
x_data = np.array(x)
y_data = np.array(y)

#  a  b   .
a = 0
b = 0

# .
lr = 0.03 

#   .
epochs = 2001 

#  .
for i in range(epochs): # epoch   
    y_hat = a * x_data + b #y   
    error = y_data - y_hat #  .
    a_diff = -(2/len(x_data)) * sum(x_data * (error)) #  a  . 
    b_diff = -(2/len(x_data)) * sum(error) #  b  .
    a = a - lr * a_diff #    a .
    b = b - lr * b_diff #    b .
    if i % 100 == 0: # 100    a, b .
        print("epoch=%.f, =%.04f, =%.04f" % (i, a, b))
        
#        .
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
