```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston_df
```

```python
boston_df.head()
```

```python
boston_df['PRICE'] = boston.target
```

```python
boston_df.head()
```

```python
print('보스톤 주택 가격 데이터셋 크기: ', boston_df.shape)
```

```python
boston_df.info()
```

```python
from sklearn.linear_model import LinearRegression #선형 회귀
from sklearn.model_selection import train_test_split #훈련 데이터와 테스트 데이터로 나누기
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#X, Y 분할하기
Y = boston_df['PRICE'] #target data로 PRICE를 Y값으로 분할
X = boston_df.drop(['PRICE'], axis = 1, inplace = False) #PRICE를 drop
```

```python
#훈련용 데이터와 평가용 데이터 분할하기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 156) 
#test_size = 0.3 전체 데이터중 테스트 데이터의 비울
```

```python
#선형 회귀 분석 : 모델 생성
lr = LinearRegression()

#선형 회귀 분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index = X.columns)
coef.sort_values(ascending = False)
```

```python
#선형 회귀 분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict))) #직선이 점들의 대표성을 갖는 비율 
```

```python
# Backward Elimination //연관성이 떨어지는 변수를 제거
import statsmodels.api as sma 
```

```python
X_train
```

```python
X_train2=sma.add_constant(X_train) #const 추가, 상수를 관찰하기위한
X_test2=sma.add_constant(X_test)
X_train2
```

```python
lr=sma.OLS(Y_train, X_train2) #어떤 변수가 필요한지....
result=lr.fit()
result.summary() #P>|t| 어떤변수가 무의미함을 알려준다 확률이 큰값 AGE... 0으로 표시된것은 중요한 변수
```

```python
X_train3=X_train2.drop('AGE', axis='columns')
X_test3=X_test2.drop('AGE', axis='columns')
X_train3
```

```python
lr=sma.OLS(Y_train, X_train3)
result=lr.fit()
result.summary()
```

```python
X_train4=X_train3.drop('INDUS', axis='columns')
X_test4=X_test3.drop('INDUS', axis='columns')
X_train4
```

```python
lr=sma.OLS(Y_train, X_train4)
result=lr.fit()
result.summary()
```

```python
Y_predict = result.predict(X_test4)

mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```
